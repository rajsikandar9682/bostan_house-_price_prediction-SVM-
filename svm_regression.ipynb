```python
# let's import  liberaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# let's create the freature and target
x=df.drop("MEDV",axis=1)
y=df["MEDV"]
```

```python
# let's splilt the data
from sklearn.model_selection import train_test_split
```

```python
# seperate data into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=20)
```

```python
# let's perform model selection
from sklearn.svm import SVR
```

```python
# create object for Sapport vactor regression
model=SVR(kernel="rbf")
```

```python
# let's feet the data
model.fit(x_train,y_train)
```

```python
# let's perform the train and test prediction
train_predict=model.predict(x_train)
test_predict=model.predict(x_test)
```

```python
# let's perform model evaluation
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2_score(y_train,train_predict)
```

```python
mean_absolute_error(y_train,train_predict)
```
